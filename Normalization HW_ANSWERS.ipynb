<H4>3 - Provide a list of employees in last name, first name order</H4>

```python
import sqlite3

conn2 = sqlite3.connect("ABC_corp.db")
curr2 = conn2.cursor()
sql = '''SELECT last_name || ', ' || first_name as name
FROM employees ORDER BY name ASC;'''
print(sql)
curr2.execute(sql)
rows = curr2.fetchall()
print("*"*35)
for row in rows:
    print(row[0]) 
conn2.close()
```

<h4>4 - Provide a list of employees in tenure order (active employees only) in descending order</h4>

```python
import sqlite3

conn2 = sqlite3.connect("ABC_corp.db")
curr2 = conn2.cursor()
sql = '''SELECT last_name || ', ' || first_name as name,
SUBSTR(hire_date,1,4)||substr(hire_date,6,2)||substr(hire_date,9,2) as hd
FROM employees ORDER BY hd ASC;'''
print(sql)
curr2.execute(sql)
rows = curr2.fetchall()
print("*"*35)
for row in rows:
    print(row[0],row[1]) 
conn2.close()
```

<h4>5 - Provide a list of employees by manager name in alphabetical order by last name</h4>

```python
import sqlite3

conn2 = sqlite3.connect("ABC_corp.db")
curr2 = conn2.cursor()
sql = '''SELECT
(SELECT B.last_name || ', ' || B.first_name FROM employees B WHERE A.manager_id = B.emp_id ) as manager ,
A.last_name || ', ' || A.first_name as name
FROM employees A ORDER BY manager ASC;'''
print(sql)
curr2.execute(sql)
rows = curr2.fetchall()
print("*"*35)
for row in rows:
    print(row[0],row[1]) 
conn2.close()
```

<h4>6 - Provide a list of employees by region</h4>

```python
import sqlite3

conn2 = sqlite3.connect("ABC_corp.db")
curr2 = conn2.cursor()
sql = '''SELECT B.description,A.last_name, A.first_name 
FROM employees A 
LEFT JOIN regions B ON B.code = A.region_code 
ORDER BY B.description,A.last_name DESC'''
print(sql)
curr2.execute(sql)
rows = curr2.fetchall()
print("*"*35)
for row in rows:
    print(row[0],row[1]) 
conn2.close()
```

<h4>7 - Provide a list of employees who have exceeded their allotment of sick days</h4>

```python
import sqlite3

conn2 = sqlite3.connect("ABC_corp.db")
curr2 = conn2.cursor()
sql = '''SELECT A.last_name, A.first_name, A.sick_days_accumulated 
FROM employees A WHERE A.sick_days_accumulated < 
(SELECT count(sick_day) FROM sick_days B WHERE A.emp_id = B.emp_id )'''
print(sql)
curr2.execute(sql)
rows = curr2.fetchall()
print("*"*35)
for row in rows:
    print(row[0],row[1],row[2]) 
conn2.close()
```
